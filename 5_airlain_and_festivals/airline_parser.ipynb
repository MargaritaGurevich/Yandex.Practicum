**Парсер для сбора с сайта данных о 11 крупнейших фестивалях 2018 года. Данные сохранены в датафрейм festivals и выведены на экран.**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL ='https://code.s3.yandex.net/learning-materials/data-analyst/festival_news/index.html'
req  = requests.get(URL)
soup=BeautifulSoup(req.text, 'lxml')
table = soup.find('table',attrs={"id": "best_festivals"})
```

```python
heads=[]
for row in table.find_all('th'):
    heads.append(row.text)  

rows=[]
for row in table.find_all('td'):
    rows.append(row.text)   

list1=[]
name=[]
place=[]
date=[]

for i in range(0, len(rows), 3):
    list1.append(rows[i:i +3])    

for j in list1:
    name.append(j[0])
    place.append(j[1])
    date.append(j[2])

festivals = pd.DataFrame({heads[0]: name,heads[1]:place, heads[2]:date})
```
